## 2. Library Imports and Configuration


```python
# Basic setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Style settings
sns.set(style="whitegrid", palette="muted", font_scale=1.1)
plt.rcParams["figure.figsize"] = (10, 6)
```

```python
# Load Dataset
books = pd.read_csv("datasets/goodreads_books_dataset.csv")

# Preview
books.head()
```

```python
# Basic information
books.info()
books.describe(include="all").T
```

```python
# Missing values
books.isnull().sum().sort_values(ascending=False)
```

## Observations ##

The dataset contains 20 variables across 3,045 records, representing metadata and rating information for books listed on Goodreads.

## 1. Variable Types

- Numerical (9): rank, percentile_rank, book_id, rating, title_length, word_count, author_count, author_name_length, series_number. These variables will primarily be used for quantitative analysis, correlation checks, and model input features.

- Categorical (7): title, author, rating_category, rating_tier, title_complexity, title_type, estimated_popularity. These variables capture descriptive and qualitative aspects such as book type, complexity, and author identity.

- Boolean (4): is_high_rated, has_series_info, has_subtitle, has_middle_name. These act as binary indicators that can be easily used in logistic models or exploratory comparisons.

## 2. Missing Values
- series_number has 1851 missing values (~61%). This likely reflects that the majority of books are standalone titles rather than part of a series. The missingness is structural, not random, meaning imputing with a neutral value (like 0) or encoding ‚Äúno series‚Äù as a separate category would be logical.

- rating_category has 2 missing values. This is negligible and likely due to a data entry or extraction issue. Dropping or filling these two values with an appropriate category (based on rating) will have minimal impact.

## 3. Initial Impressions
Overall, dataset is well-strcutured, with clear separation between numeric, categorical and boolean  fields. With both quantitative and qualitative variables, it makes it suitable for both statistical modeling and pattern analysis. The variables are intuitively meaningful, for instance title_length, title_complexity, and has_subtitle may influence readability or perceived quality, making them useful predictors in later models.

## 4. Next Steps

- Decide how to handle missing series_number values(likely encoding non-series as 0)

- Prepare categorical and boolean variables for encoding to support regression analysis.


## Univariate Analysis (Single variable exploration)

```python
# Distribution of Ratings
sns.histplot(books["rating"], bins=30, kde=True)
plt.title("Distribution of Book Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
# Title Length Distribution
sns.histplot(books["title_length"], bins=30, kde=True, color="orange")
plt.title("Distribution of Title Lengths")
plt.xlabel("Title Length (characters)")
plt.ylabel("Count")
plt.show()
```

```python
# Rating Category Counts
sns.countplot(data=books, x="rating_category", order=books["rating_category"].value_counts().index)
plt.title("Distribution of Rating Categories")
plt.xlabel("Rating Category")
plt.ylabel("Number of Books")
plt.show()
```

## Univariate Analysis: Observations

- Book Ratings: The rating distribution is slightly left-skewed (mean = 4.06 < median = 4.07), showing that most books receive high ratings, with a small number of low-rated outliers. The first quartile (3.89) indicates that 75% of books are rated above 3.89, confirming a concentration of strong ratings.

- Title Lengths: Title lengths are right-skewed (mean = 32.6 > median = 30). A handful of exceptionally long titles (up to 166 characters) extend the right tail of the distribution, pulling the mean upward. These likely correspond to books with complex titles or long subtitles.

- Rating Categories: The dataset is heavily dominated by ‚ÄúExcellent‚Äù (60%) and ‚ÄúGood‚Äù (40%) ratings, with only three entries in the ‚ÄúFair‚Äù or ‚ÄúPoor‚Äù categories. This suggests the data primarily represents high-rated or curated books, which should be considered when interpreting overall rating trends or building predictive models.

```python
# Top 10 Authors by Count
top_authors = books["author"].value_counts().head(10)
sns.barplot(x=top_authors.values, y=top_authors.index, orient='h')
plt.title("Top 10 Most Frequent Authors")
plt.xlabel("Number of Books")
plt.ylabel("Author")
plt.show()
```

```python
# Relationship Between Title Length and Rating
sns.scatterplot(data=books, x="title_length", y="rating", hue="estimated_popularity")
plt.title("Title Length vs Rating")
plt.xlabel("Title Length (characters)")
plt.ylabel("Rating")
plt.show()
```

```python
# Series Info vs Rating
sns.boxplot(data=books, x="has_series_info", y="rating")
plt.title("Do Series Books Get Higher Ratings?")
plt.xlabel("Has Series Info")
plt.ylabel("Rating")
plt.show()
```

```python
# Rating Category Counts
sns.countplot(data=books, x="rating_category", order=books["rating_category"].value_counts().index)
plt.title("Distribution of Rating Categories")
plt.xlabel("Rating Category")
plt.ylabel("Number of Books")
plt.show()
```

```python
# Correlation Heatmap (numeric features)
numeric_books = books.select_dtypes(include=["int64", "float64"])
corr = numeric_books.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap="coolwarm", center=0)
plt.title("Correlation Heatmap of Numerical Features")
plt.show()
```

```python
# Popularity by Rating Tier
sns.boxplot(data=books, x="rating_tier", y="rating", hue="estimated_popularity")
plt.title("Rating by Tier and Estimated Popularity")
plt.xticks(rotation=45)
plt.show()
```
